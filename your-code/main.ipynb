## Challenge 1
There is a box with 10 white balls, 12 red balls and 8 black balls. Calculate the probability of:
* Taking a white ball out.
* Taking a white ball on first extraction and a black ball out on second extraction without replacing the first one.
* Taking a white ball on first extraction and a black ball out on second extraction replacing the first one.
* Taking 3 balls and them being all different colours.

```python
# YOUR CODE HERE
box = []
#add 10 white balls, 12 red balls and 8 black balls
box += ['white'] * 10
box += ['red']   * 12
box += ['black'] * 8

#count each set
white_balls = box.count('white')
black_balls = box.count('black')
red_balls = box.count('red')
total_balls = len(box)

#calculate probability of picking a white ball
p_white = white_balls / total_balls
print(f"probability to:\nTake a white ball: {round(p_white*100,2)}%")

#update the state
#remaining_white = white_balls - 1
remaining_totals = total_balls -1

#calculate the probability of picking a black ball after a white ball without replacing
p_black_second_given_white_first_without_replacement = black_balls/(remaining_totals)
p_white_black_no_replacement = p_white * p_black_second_given_white_first_without_replacement
print(f"probability to:\nTake a white ball and then a black ball without replacing: {round(p_white_black_no_replacement*100,2)}%")

#calculate the probability of picking a black ball after a white ball with replacing
p_black_second_given_white_first = black_balls/(total_balls)
p_white_black_replacement = p_white * p_black_second_given_white_first
print(f"probability to:\nTake a white ball on first extraction and a black ball out on second extraction replacing the first one: {round(p_white_black_replacement*100,2)}%")


```

## Challenge 2

You want to go on picnic on Saturday. You hate rain, and would know to plan your weekend. To help you make a decision, you gather the following data about rainy days:

* 40% of all days are cloudy.
* About 50% of cloudy days are rainy. 
* About 20% of non-cloudy days are rainy.

What is the chance of rain during next Saturday?

```python
# YOUR CODE HERE
cloudy = 0.4
non_cloudy = 0.6
cloudy_rainy = cloudy * 0.5
non_cloudy_rainy = non_cloudy * 0.2

#rainy = cloudy_rainy * non_cloudy_rainy 
'''WRONG: here you are calculating the probability both things happen, which is impossible
The real question should be what's the chance rain happens in either cloudy or non cloudy weather'''
```

C = cloudy
N = not cloudy
R = rainy

What is P(R)?

Rain can happen in two separate worlds:
- it’s cloudy and rainy → C∩R
- it’s not cloudy and rainy → ¬N∩R

So in set terms:
R=(C∩R)∪(N∩R)

this means they are **disjoint**:

P(A∪B)=P(A)+P(B) **----->** P(R)=P(C∩R)+P(N∩R)





```python
rainy = cloudy_rainy + non_cloudy_rainy 
print(f"It's going to rain with a probability of {round(rainy*100,2)}%")
```
